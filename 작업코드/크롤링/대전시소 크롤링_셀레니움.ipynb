```python
from selenium import webdriver
import re
import pandas as pd
import time
import math
```

```python
#크롬브라우저 열기
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.daejeon.go.kr/seesaw/djaskView.do?sugstSeq=2272&menuSeq=5923')
driver.implicitly_wait(10)  #페이지가 열릴때까지 대기
```

```python
result = [  ]

#댓글 데이터 가져오기
def get_data(url):
    resp = requests.get(driver.current_url)
    html = BeautifulSoup(resp.content, 'html.parser')
    reply_lst = html.find('ul', {'class': 'reply_lst'})
    lis = reply_lst.findAll('li')
    for li in lis:
        created_at = datetime.strptime(reply_lst.find('em').findAll('span')[0].getText(), "%Y-%m-%d %H:%M")
        reply_text = li.find('div').getText()
        like = li.find('a', {'class': 'btn_empathy up small'}).findAll('em')[0].getText()
        
        data = [created_at,like ,reply_text]
        
        if int(like) > 0:
            result.append(data)
            print(created_at,like ,reply_text)
        else:
            break
```

```python
get_data(driver.current_url)
```

```python
#총 댓글수
reply_count = html.find('div', {'class':'bd_reply'}).find('strong').getText()
reply_count = reply_count.replace("[","")
reply_count = reply_count.replace("]","")
int(reply_count)
```

```python
#다음페이지 클릭
def click_next_page(driver):
    
    #다음페이지 버튼 리스트
    page_bar = driver.find_elements_by_css_selector('#bd_reply > div')[0]

    #현재 선택되어 있는 페이지는 포함 안됨
    pages = page_bar.find_elements_by_css_selector('a')

    #현재페이지
    page_now = page_bar.find_elements_by_css_selector('strong')[0].text
    page_now
    
    int_page_now = int(page_now)
    
    #10에서 11로 넘어갈때 12에서 3으로 다시 돌아감
    if int_page_now >= 11 :
        int_page_now -= 10
    
    next_page = '#bd_reply > div > a:nth-child('+ str(int_page_now+3) + ')'

    click_next_page = driver.find_element_by_css_selector(next_page)
    click_next_page.click()
    time.sleep(3)

#페이지가 4 ~ 12인 형식, 3는 첫번째 페이지이므로 필요는 없음
#다음 페이지바 #bd_reply > div > a.direction.last  

click_next_page(driver)
```

```python
result = [  ]

for i in range(1, math.ceil(int(reply_count)/10)):
    get_data(driver.current_url)
    click_next_page(driver)
    i += 1
    
driver.close()
```

```python
data = pd.DataFrame(result)

data.head()
```
